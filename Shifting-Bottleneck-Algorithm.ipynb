# The Shifting Bottleneck Algorithm (SBA)

The Shifting Bottleneck Algorithm (SBA) is a heuristic method widely used for solving job shop scheduling problems. A job shop is a setting where jobs must be processed on one or more machines. The set of machines that a job must be processed on and the processing order can vary by job. The goal is to minimize the makespan, or the total time required to complete all jobs. The algorithm iteratively identifies and schedules "bottleneck" machines, where the bottleneck machine at any iteration is the machine that is the most constrained or likely to cause the most significant delay. When a new bottleneck machine is identified, the order of jobs on that machine is fixed before continuing to identify the next bottleneck. If at any point, there is no bottleneck machine, the jobs are set for all unsequenced machines and the algorithm stops.

## Algorithm Overview
The SBA generally follows these steps:

1. Initialize the schedule with no job-to-machine assignments.
2. Identify the bottleneck machine for the current unscheduled job assignments.
3. Schedule the bottleneck machine by optimizing the sequence of jobs on that machine.
4. Update constraints across all machines based on the bottleneck machine’s schedule, taking into account the updated flow of jobs.
5. Repeat steps 2–4 until all machines are scheduled.

The original paper that presented the algorithm (see reference below) included a *reoptimization* step that was performed after every iteration. We will ignore this step in the implementation we present in this notebook.

### Reference
Adams, J., Balas, E., & Zawack, D. (1988). "The Shifting Bottleneck Procedure for Job Shop Scheduling." Management Science, 34(3), 391-401.

## Single Pass Implementation
We will first look at how to implement one *pass* (steps 2 - 4) of the algorithm. Then, we will define functions for the various steps and combine them to create a complete version.

### Critical Path Method (CPM) Function

When executing the SBA, we will represent the job shop environment as a graph, where vertices correspond to job/machine combinations and edges are used to enforce the machine sequence for jobs (conjunctive arcs) and the job sequence for machines (disjunctive arcs). The conjunctive arcs are given from the data, i.e., from the machine sequences for each job. However, we must decide on the disjunctive arcs to add. As we add arcs, we will use the critical path method to determine the length of the longest path, i.e., the critical path, through the graph representation of the job shop scheduling instance. The following code block defines a function for determining the critical path for a set of *project* data, where the *project* in our case is the job shop scheduling instance.

```python
def execute_CPM(
    project_data_dict: dict,
    start_time: float = 0.0,
) -> dict:
    """
    The project_data_dict object should have keys that correspond to tasks.
    The values for each key should be a dictionary with three elements:
        1: time - a numeric value representing the task time
        2: predecessors - a list of tasks that must be completed before the current task starts
        3: successors - a list of tasks that cannot start until the current task completes

    Example for project_data_dict format:
    project_data_dict = {
        'A': {
            'time': 1,
            'predecessors': [],
            'successors': ['B'],
        },
        'B': {
            'time': 1,
            'predecessors': ['A'],
            'successors': [],
        },
    }
    """
    import copy

    project_data_dict = copy.deepcopy(project_data_dict)
    
    max_time = start_time + sum([val['time'] for val in project_data_dict.values()])
    for task in project_data_dict:
        project_data_dict[task]['ES'] = 0.0
        project_data_dict[task]['EF'] = 0.0
        project_data_dict[task]['LS'] = max_time
        project_data_dict[task]['LF'] = max_time

    # forward pass
    unprocessed_tasks = set(project_data_dict.keys())    
    while unprocessed_tasks:
        processed_tasks = set()
        for task in unprocessed_tasks:
            predecessors = project_data_dict[task]['predecessors']
            if predecessors:
                unprocessed_predecessors = set(project_data_dict[task]['predecessors']).intersection(unprocessed_tasks)
                if not unprocessed_predecessors:
                    max_predecessor_EF = 0
                    for predecessor in predecessors:
                        if project_data_dict[predecessor]['EF'] > max_predecessor_EF:
                            max_predecessor_EF = project_data_dict[predecessor]['EF']
                    project_data_dict[task]['ES'] = max_predecessor_EF
                    project_data_dict[task]['EF'] = max_predecessor_EF + project_data_dict[task]['time']
                    processed_tasks.add(task)
            else:
                project_data_dict[task]['ES'] = start_time
                project_data_dict[task]['EF'] = start_time + project_data_dict[task]['time']
                processed_tasks.add(task)

        if not processed_tasks:
            print(' - Foward pass iteration completed without update. Something is wrong. Check project data')
            break
        for task in processed_tasks:
            unprocessed_tasks.remove(task)
    
    # project length computation
    project_length = 0
    for task in project_data_dict:
        if project_data_dict[task]['EF'] > project_length:
            project_length =  project_data_dict[task]['EF']
    
    # backward pass
    unprocessed_tasks = set(project_data_dict.keys())
    while unprocessed_tasks:
        processed_tasks = set()
        for task in unprocessed_tasks:
            successors = project_data_dict[task]['successors']
            if successors:
                unprocessed_successors = set(project_data_dict[task]['successors']).intersection(unprocessed_tasks)
                if not unprocessed_successors:
                    min_successor_LS = project_length
                    for successor in successors:
                        if project_data_dict[successor]['LS'] < min_successor_LS:
                            min_successor_LS = project_data_dict[successor]['LS']
                    project_data_dict[task]['LF'] = min_successor_LS
                    project_data_dict[task]['LS'] = min_successor_LS - project_data_dict[task]['time']
                    processed_tasks.add(task)
            else:
                project_data_dict[task]['LF'] = project_length
                project_data_dict[task]['LS'] = project_length - project_data_dict[task]['time']
                processed_tasks.add(task)
    
        if not processed_tasks:
            print(' - Backward pass iteration completed without update. Something is wrong. Check project data')
            break
        for task in processed_tasks:
            unprocessed_tasks.remove(task)
    
    # slack computations
    for task in project_data_dict:
        project_data_dict[task]['slack'] = project_data_dict[task]['LF'] - project_data_dict[task]['EF']
    
    # determine critical activities
    crticial_activities = []
    for task in project_data_dict:
        if project_data_dict[task]['slack'] == 0:
            project_data_dict[task]['critical'] = True
            crticial_activities.append(task)
        else:
            project_data_dict[task]['critical'] = False

    return project_data_dict
```

### Ganntt Chart Function

The following code block defines a function that will allow us to visualize the critical path for a project instance after applying the CPM.

```python
def make_gantt_chart(
    project_data_dict: dict, 
    figsize: tuple = (8, 3)
):
    """
    The project_data_dict object should have keys that correspond to tasks.
    The values for each key should be a dictionary that contains the following elements:
        1: ES - early start time for task from CPM
        2: EF - early finish time for task from CPM
        3: critical - a boolean value that indicates whether or not the activity is on a critical path

    Example for project_data_dict format:
    project_data_dict = {
        'A': {
            'time': 1,
            'predecessors': [],
            'successors': ['B'],
            'ES': 0,
            'EF': 1,
            'LS': 0,
            'LF': 1,
            'critical': True,
        },
        'B': {
            'time': 1,
            'predecessors': ['A'],
            'successors': [],
            'ES': 1,
            'EF': 2,
            'LS': 1,
            'LF': 2,
            'critical': True,
        }
    }
    """
    import matplotlib.pyplot as plt
    
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    
    for task in project_data_dict:
        start_time = project_data_dict[task]['ES']
        completion_time = project_data_dict[task]['EF']
        length = completion_time - start_time
        if project_data_dict[task]['critical']:
            task_color='red'
        else:
            task_color='lightgray'
        ax.barh(
            task, 
            length,
            left=start_time,
            color=task_color,
            edgecolor='k',
        )
        ax.text(
            start_time + length/2,
            task,
            task,
            va='center',
            ha='center',
            color='k',
        )
    ax.spines[['right', 'top']].set_visible(False)
    
    ax.set_xlabel('Time')
    ax.set_ylabel('Task')
    ax.set_title('Gantt Chart')
    plt.show()
```

### Instance Data

The following code block defines the data for job shop scheduling instance we will test with.

```python
job_shop_instance = {
    'J1': {
        'machine_sequence': ['M1', 'M2', 'M3'],
        'processing_times': {'M1': 5, 'M2': 10, 'M3': 8},
    },
    'J2': {
        'machine_sequence': ['M2', 'M1', 'M3'],
        'processing_times': {'M1': 7, 'M2': 6, 'M3': 9},
    },
    'J3': {
        'machine_sequence': ['M3', 'M1', 'M2'],
        'processing_times': {'M1': 8, 'M2': 4, 'M3': 12},
    },
    'J4': {
        'machine_sequence': ['M1', 'M3', 'M2'],
        'processing_times': {'M1': 6, 'M2': 9, 'M3': 5},
    },
}

machines = []
for job in job_shop_instance:
    for machine in job_shop_instance[job]['machine_sequence']:
        machines.append(machine)
machines = sorted(list(set(machines)))
```

The following code block converts the instance data into a representation that is compatible with our CPM implementation.

```python
network_data = {}
for job, job_data in job_shop_instance.items():
    machine_sequence = job_data['machine_sequence']
    processing_times = job_data['processing_times']
    first_machine = True
    previous_job_machine = None
    for machine in machine_sequence:
        job_machine = f'{job}-{machine}'
        if previous_job_machine:
            network_data[f'{job}-{machine}'] = {
                'predecessors': [previous_job_machine],
                'time': processing_times[machine],
            }
        else:
            network_data[f'{job}-{machine}'] = {
                'predecessors': [],
                'time': processing_times[machine],
            }
        previous_job_machine = job_machine


for job in network_data:
    network_data[job]['successors'] = []

for job in network_data:
    predecessors = network_data[job]['predecessors']
    for predecessor in predecessors:
        network_data[predecessor]['successors'].append(job)
```

The following code block applies the CPM to the data and plots the results. Note that at this point, we are not enforcing the constraint that a machine can only process a single job at any time. In particular, our *graph* representation only includes the conjunctive arcs (sequence of machines for each job).

```python
CPM_data = execute_CPM(network_data)

make_gantt_chart(CPM_data)
```

### Identifying the *Bottleneck* Machine

The following code block implements a procedure to identify the first *bottleneck* machine. This process consists of formulating the problem for determining the sequence of jobs on all unscheduled machines as an instance of $1/r_{j}/L_{max}$, where:
 - $r_{j} = ES_{j}$, and
 - $d_{j} = LF_{j}$.

In the implementation presented in the original paper, these instances are solved to optimality. However, in this implementation, I simply use a static dispatching rule that: "Schedules the available job with the *earliest due date* whenever the machine is free,", i.e., an earliest due date or EDD rule.

```python
bottleneck_computations = {}
for machine in machines:
    machine_data = []
    for job_machine, job_machine_CPM_Data in CPM_data.items():
        if job_machine.endswith(machine):
            machine_data.append({
                'job': job_machine,
                'pj': job_machine_CPM_Data['time'],
                'rj': job_machine_CPM_Data['ES'],
                'dj': job_machine_CPM_Data['LF'],
            })
    
    machine_data = pd.DataFrame(machine_data)
    machine_data = machine_data.set_index('job')
    machine_data = machine_data.sort_values(['rj', 'dj'])
    
    unscheduled_jobs = set(machine_data.index.to_list())
    unscheduled_job_data = machine_data.copy()
    
    schedule = []
    t = unscheduled_job_data['rj'].min()
    
    while unscheduled_jobs:
        available_job_filter = unscheduled_job_data['rj'] <= t
        available_job_data = unscheduled_job_data[available_job_filter]
        if len(available_job_data) > 0:
            job_to_schedule = available_job_data['dj'].idxmin()
            t += available_job_data.loc[job_to_schedule, 'pj']
            schedule.append(job_to_schedule)
            unscheduled_jobs.remove(job_to_schedule)
            unscheduled_job_data = machine_data[machine_data.index.isin(unscheduled_jobs)]
        else:
            t = unscheduled_job_data['rj'].min()
    
    t = 0
    Lmax = -np.inf
    for job in schedule:
        if machine_data.loc[job, 'rj'] > t:
            t = machine_data.loc[job, 'rj']
        completion_time = t + machine_data.loc[job, 'rj']
        lateness = completion_time - machine_data.loc[job, 'dj']
        t = completion_time
        if lateness > Lmax:
            Lmax = lateness
    print(f' - {machine}: {Lmax}')

    bottleneck_computations[machine] = {
        'sequence': schedule,
        'Lmax': Lmax,
    }

sorted_bottleneck_computations = sorted(
    bottleneck_computations.items(), 
    key=lambda x: x[1]['Lmax'], 
    reverse=True,
)

bottleneck_machine = sorted_bottleneck_computations[0][0]
bottleneck_sequence = bottleneck_computations[bottleneck_machine]['sequence']

print(f' - {bottleneck_machine = }')
print(f' - {bottleneck_sequence = }')
```

### Add Disjunctive (Job Sequence for Machine) Arcs for Bottleneck Machine

```python
for job1, job2 in zip(bottleneck_sequence[:-1], bottleneck_sequence[1:]):
    
    if job1 not in network_data[job2]['predecessors']:
        network_data[job2]['predecessors'].append(job1)
    
    if job2 not in network_data[job1]['successors']:
        network_data[job1]['successors'].append(job2)
```

Rerun CPM for Updated Graph

```python
CPM_data = execute_CPM(network_data)

make_gantt_chart(CPM_data)
```

## End-to-End Implementation

We will now implement the procedure end-to-end. The following code block implements our function for performing the bottleneck computations for a set of unscheduled machines.

```python
def get_bottleneck_computations(
    CMP_data_dict: dict,
    unscheduled_machines: list,
) -> dict:
    
    bottleneck_computations = {}
    for machine in unscheduled_machines:
        machine_data = []
        for job_machine, job_machine_CPM_Data in CMP_data_dict.items():
            if job_machine.endswith(machine):
                machine_data.append({
                    'job': job_machine,
                    'pj': job_machine_CPM_Data['time'],
                    'rj': job_machine_CPM_Data['ES'],
                    'dj': job_machine_CPM_Data['LF'],
                })
        
        machine_data = pd.DataFrame(machine_data)
        machine_data = machine_data.set_index('job')
        machine_data = machine_data.sort_values(['rj', 'dj'])
        
        unscheduled_jobs = set(machine_data.index.to_list())
        unscheduled_job_data = machine_data.copy()
        
        schedule = []
        t = unscheduled_job_data['rj'].min()
        
        while unscheduled_jobs:
            available_job_filter = unscheduled_job_data['rj'] <= t
            available_job_data = unscheduled_job_data[available_job_filter]
            if len(available_job_data) > 0:
                job_to_schedule = available_job_data['dj'].idxmin()
                t += available_job_data.loc[job_to_schedule, 'pj']
                schedule.append(job_to_schedule)
                unscheduled_jobs.remove(job_to_schedule)
                unscheduled_job_data = machine_data[machine_data.index.isin(unscheduled_jobs)]
            else:
                t = unscheduled_job_data['rj'].min()
        
        t = 0
        Lmax = -np.inf
        for job in schedule:
            if machine_data.loc[job, 'rj'] > t:
                t = machine_data.loc[job, 'rj']
            completion_time = t + machine_data.loc[job, 'rj']
            lateness = completion_time - machine_data.loc[job, 'dj']
            t = completion_time
            if lateness > Lmax:
                Lmax = lateness
    
        bottleneck_computations[machine] = {
            'sequence': schedule,
            'Lmax': Lmax,
        }

    return bottleneck_computations
```

The following code block includes the end-to-end implementation

```python
print(' - Defining data')
job_shop_instance = {
    'J1': {
        'machine_sequence': ['M1', 'M2', 'M3'],
        'processing_times': {'M1': 5, 'M2': 10, 'M3': 8},
    },
    'J2': {
        'machine_sequence': ['M2', 'M1', 'M3'],
        'processing_times': {'M1': 7, 'M2': 6, 'M3': 9},
    },
    'J3': {
        'machine_sequence': ['M3', 'M1', 'M2'],
        'processing_times': {'M1': 8, 'M2': 4, 'M3': 12},
    },
    'J4': {
        'machine_sequence': ['M1', 'M3', 'M2'],
        'processing_times': {'M1': 6, 'M2': 9, 'M3': 5},
    },
}

print(' - Getting list of machines')
machines = []
for job in job_shop_instance:
    for machine in job_shop_instance[job]['machine_sequence']:
        machines.append(machine)
machines = sorted(list(set(machines)))

print(' - Getting initial graph representation')
network_data = {}
for job, job_data in job_shop_instance.items():
    machine_sequence = job_data['machine_sequence']
    processing_times = job_data['processing_times']
    first_machine = True
    previous_job_machine = None
    for machine in machine_sequence:
        job_machine = f'{job}-{machine}'
        if previous_job_machine:
            network_data[f'{job}-{machine}'] = {
                'predecessors': [previous_job_machine],
                'time': processing_times[machine],
            }
        else:
            network_data[f'{job}-{machine}'] = {
                'predecessors': [],
                'time': processing_times[machine],
            }
        previous_job_machine = job_machine


for job in network_data:
    network_data[job]['successors'] = []

for job in network_data:
    predecessors = network_data[job]['predecessors']
    for predecessor in predecessors:
        network_data[predecessor]['successors'].append(job)


print(' - SBA Initialization')
CPM_data = execute_CPM(network_data)
unscheduled_machines = list(machines)
iteration = 0
print('-'*40)

while unscheduled_machines:
    iteration += 1
    print(f'   - Starting SBA iteration {iteration}')
    
    bottleneck_computations = get_bottleneck_computations(
        CMP_data_dict = CPM_data,
        unscheduled_machines=unscheduled_machines,
    )
    
    sorted_bottleneck_computations = sorted(
        bottleneck_computations.items(), 
        key=lambda x: x[1]['Lmax'], 
        reverse=True,
    )
    
    bottleneck_machine = sorted_bottleneck_computations[0][0]
    bottleneck_sequence = bottleneck_computations[bottleneck_machine]['sequence']
    unscheduled_machines.remove(bottleneck_machine)
    
    print(f'   - {bottleneck_machine = }')
    print(f'   - {bottleneck_sequence = }')
    
    for job1, job2 in zip(bottleneck_sequence[:-1], bottleneck_sequence[1:]):
        
        if job1 not in network_data[job2]['predecessors']:
            network_data[job2]['predecessors'].append(job1)
        
        if job2 not in network_data[job1]['successors']:
            network_data[job1]['successors'].append(job2)
    
    CPM_data = execute_CPM(network_data)
    print('-'*40)

print(' - Final Schedule')
make_gantt_chart(CPM_data)
```
